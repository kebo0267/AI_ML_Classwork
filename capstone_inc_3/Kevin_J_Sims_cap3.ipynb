```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NSMES1988.csv")
```

```python
df.head(100)
```

```python
df.info()
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
df[numeric_cols].max()
```

```python
numeric_cols = df.select_dtypes(include=['int']).columns.tolist()
float_cols = df.select_dtypes(include=['float']).columns.tolist()

print(numeric_cols)
print(float_cols)
```

```python
for colname in numeric_cols:
    df[colname] = df[colname].astype("int16")

for colname in float_cols:
    df[colname] = df[colname].astype("float32")
df.info()
```

```python
df.rename(columns={"visits":"dr_visits","nvisits":"non_dr_visits","ovisits":"out_pat_dr_visits"},inplace=True)
df.rename(columns={"novisits":"out_pat_non_dr_visits","emergency":"er_visits","health":"self_health_assessment"},inplace=True)
```

```python
obj_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
for column_name in obj_cols:
    if df[column_name].nunique() == 2 and column_name != "gender":
        df[column_name] = df[column_name].replace({"yes":True,"no":False,"normal":False,"limited":True})
        df[column_name] = df[column_name].astype("bool")
        
df.head(100)
df.info()
```

1. Recommend renaming the columns to more descriptive title.
2. Based on max values for columns of type integer, all integer columns should be resized to in16
3. Based on max values for columns of type float, all float columns should be resized to float32
4. Change columns with 2 unique values to boolean types.  Note: Gender while in the data set is expressed as a binary choice, because of the floating political views of gender will be maintained as an object.

```python
#Correct bad income data.  Any data <0 is set to 0.
df['income'] = df['income'].where(df['income'] >= 0, 0)
```

```python
#Visit stats
visits = ["dr_visits","non_dr_visits","out_pat_dr_visits","out_pat_dr_visits","er_visits","er_visits","hospital"]
agg_1 = df[visits].agg(["mean","median"])

print(agg_1.transpose())
```

```python
cat_stats = ["gender","married","employed","insurance","medicaid"]
for cat in cat_stats:
    value_cnt = df[cat].value_counts()
    print(value_cnt)
    print()
```

```python
#Education level and Income Stats
linear_data = ["school","income"]
agg_1 = df[linear_data].agg(["mean","median"])

print(agg_1.transpose())
```

```python
# Distribution of age separated by gender
age_gen_stats = df.groupby(["age","gender"]).size().unstack(fill_value=0)
print(age_gen_stats)

print(df.groupby("gender")["age"].agg(["min","mean","max"]))
```

```python
# Distribution of health status separated by age
age_health_stats = df.groupby(["age","self_health_assessment"]).size().unstack(fill_value=0)
print(age_health_stats)
```

```python
# Distribution of income by gender
income_gender_stats = df.groupby("gender")["income"].agg(["min","mean","max"])
print(income_gender_stats)
```

```python
# Distribution of income by region
region_income_stats = df.groupby("region")["income"].agg(["min","mean","max"])
print(region_income_stats)
```

```python
# Distribution of income by age
age_income_stats = df.groupby("age")["income"].agg(["min","mean","max"])
print(age_income_stats)
```
